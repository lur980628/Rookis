```python
# 필요한 라이브러리를 임포트합니다.
# requests는 웹 요청을 위해 필요합니다.
# os는 파일 및 디렉터리 경로를 다루기 위해 필요합니다.
import requests
import json
import os

# 스크래핑할 웹사이트의 기본 URL을 정의합니다.
base_url = 'https://www.pimfyvirus.com/dataurl_action/homeprotectionlist/p=/search/01/p={}//'

# HTTP 요청 헤더를 정의합니다.
# "X-Requested-With": "XMLHttpRequest"는 JSON 응답을 얻는 데 필요합니다.
req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36",
    "X-Requested-With": "XMLHttpRequest"
}

# 파일 경로를 정의합니다.
output_dir = "data"
output_file = os.path.join(output_dir, "dog_id.txt")

# 'data' 디렉터리가 없으면 생성합니다.
# 이렇게 하면 'FileNotFoundError'를 방지할 수 있습니다.
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"'{output_dir}' 디렉터리가 생성되었습니다.")
else:
    print(f"'{output_dir}' 디렉터리가 이미 존재합니다.")

# 파일을 쓰기 모드('w')로 엽니다.
with open(output_file, "w", encoding="utf-8") as f:
    # 1부터 185까지 페이지를 반복합니다.
    for i in range(1, 186):
        url = base_url.format(i)

        try:
            # POST 요청을 보냅니다.
            res = requests.post(url, headers=req_header)
            
            # 요청이 성공했는지 확인합니다.
            if res.ok:
                print(f"성공적으로 페이지 {i}에 접속했습니다: {res.url}")
                # 응답을 JSON 형식으로 파싱합니다.
                data = res.json()
                
                # 'lst' 키를 가진 리스트를 가져옵니다.
                for item in data.get('lst', []):
                    tags = item.get('tag', [])
                    link = item.get('wish', '')
                    
                    # 태그에 '임보가능', '임보중', '입양전제', '렐레이임보'가 포함되어 있는지 확인합니다.
                    for tag in tags:
                        if tag in ['임보가능', '임보중', '입양전제', '렐레이임보']:
                            # 링크에서 강아지 ID를 추출합니다.
                            dog_id = link.strip('/').split('/')[-1]
                            # 추출한 ID를 파일에 씁니다.
                            f.write(dog_id + '\n')
                            print(f'ID 저장됨: {dog_id} (태그: {tags}, 링크: {link})')
                            # 조건에 맞는 첫 번째 태그를 찾으면 루프를 종료합니다.
                            break
            else:
                # 요청 실패 시 오류 코드를 출력합니다.
                print(f"페이지 {i} 요청 실패: Error Code = {res.status_code}")
        except requests.exceptions.RequestException as e:
            # 네트워크 오류 등 예외 발생 시 출력합니다.
            print(f"요청 중 오류 발생: {e}")
            continue

print("모든 강아지 ID 스크래핑 및 파일 저장이 완료되었습니다.")
```
