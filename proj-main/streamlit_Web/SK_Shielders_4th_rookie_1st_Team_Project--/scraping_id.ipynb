```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
# 강아지 
import json

base_url= 'https://www.pimfyvirus.com/dataurl_action/homeprotectionlist/p=/search/01/p={}//'
# print(type(base_url))


req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36",
    "X-Requested-With": "XMLHttpRequest" #파일구조가 Json이라 필요 
}

with open("data/dog_id.txt", "w", encoding="utf-8") as f:
    for  i in range(1,186): #페이지 범위 
        url=base_url.format(i)

        #network에서 응답이 post로 처리하게 되어있음 
        res=requests.post(url,headers=req_header)
    
        if(res.ok):
            print(res.url)
            data = res.json()
            
            for item in data.get('lst',[]):
                tags=item['tag']
                link = item['wish']
                for tag in tags: 
                    if tag in ['임보가능','임보중','입양전제', '렐레이임보']:
                        dog_id = link.strip('/').split('/')[-1]  # /dataurl_action/homeprotectionwish/2955
                        f.write(dog_id + '\n')
                        print(f'{tags} {link}') 
                        break  
                       
        else:
            print(f'Error Code = {res.status_code}') 

```
